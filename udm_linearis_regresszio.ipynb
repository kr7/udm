```python
import matplotlib.pyplot as plt
import numpy as np
```

**1. Egyváltozós lineáris regresszió, y = wx alakú modell, konkrét adatokon**

```python
x = [26, 43, 60]
```

```python
y = [35, 50, 75]
```

```python
all_w = []
eps = 0.0001
w = 10
for i in range(20):
  w = w - eps * (1/3)*( 2*(26*w-35)*26 + 2*(43*w-50)*43 + 2*(60*w-75)*60 )
  print(f"{i:3} {w}")
  all_w.append(w)
```

```python
plt.xticks(range(0, 20, 5))
plt.plot(all_w)
plt.show()
```

**2. Egyváltozós lineáris regresszió, y = wx alakú modell, tetszőleges adatokon**

```python
all_w = []
eps = 0.0001
w = 10
for i in range(20):
  hibafuggveny_derivaltja = 0
  for j in range(len(x)):
    hibafuggveny_derivaltja += 2*(w*x[j]-y[j])*x[j]
  hibafuggveny_derivaltja = hibafuggveny_derivaltja/len(x)
  w = w - eps * hibafuggveny_derivaltja
  print(f"{i:3} {w}")
  all_w.append(w)
```

```python
plt.xticks(range(0, 20, 5))
plt.plot(all_w)
plt.show()
```

**3. Kétváltozós lineáris regresszió, $y = w^{(1)}*x^{(1)} + w^{(2)}*x^{(2)}$ alakú modell**

```python
# Adatok
x = [ [26, 1], [40, 2], [59, 3], [45, 2] ]
y = [ 30, 45, 60, 50 ]
```

```python
eps = 0.0001
w1 = 0
w2 = 0
for i in range(20):
  hibafuggveny_w1_szerinti_parcialis_derivaltja = 0
  hibafuggveny_w2_szerinti_parcialis_derivaltja = 0
  for j in range(len(x)):
    pred = w1*x[j][0] + w2*x[j][1]
    hibafuggveny_w1_szerinti_parcialis_derivaltja += 2*x[j][0]*(pred-y[j])
    hibafuggveny_w2_szerinti_parcialis_derivaltja += 2*x[j][1]*(pred-y[j])
  hibafuggveny_w1_szerinti_parcialis_derivaltja = hibafuggveny_w1_szerinti_parcialis_derivaltja/len(x)
  hibafuggveny_w2_szerinti_parcialis_derivaltja = hibafuggveny_w2_szerinti_parcialis_derivaltja/len(x)

  w1 = w1 - eps * hibafuggveny_w1_szerinti_parcialis_derivaltja
  w2 = w2 - eps * hibafuggveny_w2_szerinti_parcialis_derivaltja
```

```python
all_w1 = []
all_w2 = []

eps = 0.0001
w1 = 0
w2 = 0
for i in range(20):
  hibafuggveny_w1_szerinti_parcialis_derivaltja = 0
  hibafuggveny_w2_szerinti_parcialis_derivaltja = 0
  for j in range(len(x)):
    pred = w1*x[j][0] + w2*x[j][1]
    hibafuggveny_w1_szerinti_parcialis_derivaltja += 2*x[j][0]*(pred-y[j])
    hibafuggveny_w2_szerinti_parcialis_derivaltja += 2*x[j][1]*(pred-y[j])
  hibafuggveny_w1_szerinti_parcialis_derivaltja = hibafuggveny_w1_szerinti_parcialis_derivaltja/len(x)
  hibafuggveny_w2_szerinti_parcialis_derivaltja = hibafuggveny_w2_szerinti_parcialis_derivaltja/len(x)

  w1 = w1 - eps * hibafuggveny_w1_szerinti_parcialis_derivaltja
  w2 = w2 - eps * hibafuggveny_w2_szerinti_parcialis_derivaltja

  print(f"{i:3} {w1:6.4f} {w2:6.4f} ")
  all_w1.append(w1)
  all_w2.append(w2)
```

```python
plt.xticks(range(0, 20, 5))
plt.plot(all_w1, label = "w1")
plt.plot(all_w2, label = "w2")
plt.legend()
plt.show()
```

**4. Lineáris regresszio tetszőlegesen sok változóval**

```python
# Adatok
x = np.array([ [26, 1, 0.5],
               [40, 2, 1.2],
               [59, 3, 1.5],
               [45, 2, 1.3],
               [30, 1, 3.5],
               [100, 5, 4.2] ])
y = np.array([ 30, 45, 60, 50, 20, 85 ])
```

```python
eps = 0.0001
w = np.zeros( len(x[0]) )
for i in range(100):
  grad = np.zeros( len(x[0]) )
  for j in range(len(x)):
    pred = 0
    for k in range(len(x[0])):
      pred = pred + w[k]*x[j][k]
    for k in range(len(x[0])):
      grad[k] = grad[k] + 2*x[j][k]*(pred-y[j])
  grad = grad/len(x) # Elemenkenti osztas!
  w = w - eps * grad # Elemenkénti szorzás ill. kivonás!
  print(f"{i:3} {w}")
```
