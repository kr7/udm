**Távolságok koncentrációja**

Ez a notebook a nagydimenziós terekre jellemző <i>distance concentration</i> jelenséget szemlélteti.  

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.metrics.pairwise import euclidean_distances, \
     manhattan_distances, cosine_distances
```

```python
ndiff = []
for k in range(1,300):
  data = np.random.rand( 1000, k )
  distances = euclidean_distances(data)
  d_max = np.max(distances)

  # Minden pont önmagától vett távolsága nulla.
  # Azért, hogy ez ne zavarjon a minimum számításánál,
  # a főátló mentén az értékeket végtelenre állítjuk
  for i in range(len(distances)):
    distances[i][i] = np.inf
  d_min = np.min(distances)

  normalized_difference = (d_max - d_min) / d_max
  ndiff.append(normalized_difference)
```

```python
plt.plot(ndiff)
plt.show()
```
